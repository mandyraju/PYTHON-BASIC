Question 1: What is hypothesis testing in statistics?

Answer: Hypothesis testing is a statistical method used to decide if there is enough evidence
in a sample of data to support or reject a particular idea (hypothesis) about a population.

It helps answer questions like “Does this new medicine work better than the old one?” or “Is the
average height of students in this city really 170cm?”



Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

• Answer: Null Hypothesis (H0):
This is the default assumption that there is no effect, no difference, or nothing
special happening.

Example: “The mean test score of students this year is the same as last year.”

• Alternative Hypothesis (H1 or Ha):
This is what you want to prove; it suggests there is an effect or a difference.

Example: “The mean test score this year is different from last year.”

• Difference:
The null hypothesis is the starting assumption, while the alternative is what you suspect or
hope is true.



Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome
of a test.

• Answer:

• Significance level (alpha, α):

It is the threshold chosen (commonly 0.05) that tells us how much risk of a “false alarm”
(Type I error) we accept.

• Role:
• If the probability (p-value) of observing our result (or more extreme) is less than α,
we reject H0 (we say “this result is statistically significant”).
• If p-value is greater than α, we fail to reject H0 (not enough evidence).

Example:
If α = 0.05, we're okay with a 5% risk of wrongly rejecting H0




Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

Type I Error (False Positive):

Rejecting the null hypothesis when it is actually true.

Example: A medical test says you’re sick (when you’re healthy).

Type II Error (False Negative):

Failing to reject the null hypothesis when it is actually false.

Example: A test says you’re healthy (when you’re actually sick).




Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

• Answer: Z-test:

• Used when sample size is large (n > 30) and the population standard deviation
is known.

• Example: Checking if the average height in a city is 165cm, knowing population
variance.

• T-test:

• Used when sample size is small or population standard deviation is unknown.

• Example: Comparing exam scores for two small classes

Summary:

Use Z-test for large samples with known σ, and T-test for small samples or unknown σ.

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then
#plot its histogram.

#answer

import numpy as np
import matplotlib.pyplot as plt
# Binomial Distribution: n=10, p=0.5, size=1000 samples
n = 10
p = 0.5
data = np.random.binomial(n, p, 1000)
# Plot histogram
plt.hist(data, bins=range(n+2), color='skyblue', edgecolor='black',
align='left')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show
#the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3,
#50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0,
#50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


#answer
import numpy as np
from scipy.stats import norm

sample_data = [
    49.1,50.2,51.0,48.7,50.5,49.8,50.3,50.7,50.2,49.6,
    50.1,49.9,50.8,50.4,48.9,50.6,50.0,49.7,50.2,49.5,
    50.1,50.3,50.4,50.5,50.0,50.7,49.3,49.8,50.2,50.9,
    50.3,50.4,50.0,49.7,50.5,49.9
]
# Hypothesis: mean = 50
sample_mean = np.mean(sample_data)
pop_mean = 50
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Calculate Z-statistic
z = (sample_mean - pop_mean) / (sample_std/np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"Sample Mean: {sample_mean:.2f}, Z-value: {z:.2f}, p-value: {p_value:.4f}")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95%
#confidence interval for its mean. Plot the data using Matplotlib.

#answer
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate Data
data = np.random.normal(100, 15, 1000)

# Mean and CI
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)
conf_int = norm.interval(0.95, loc=mean, scale=std/np.sqrt(n))
print(f"Mean: {mean:.2f}, 95% CI: ({conf_int[0]:.2f}, {conf_int[1]:.2f})")

# Plot
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Simulated Normal Data')
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the
#standardized data using a histogram. Explain what the Z-scores represent in terms of standard
#deviations from the mean.

#answer

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, color='skyblue', edgecolor='black')
plt.title('Z-scores Histogram')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.show()
```
